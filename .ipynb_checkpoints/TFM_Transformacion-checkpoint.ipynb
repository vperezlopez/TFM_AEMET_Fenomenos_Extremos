```python
import pandas as pd
import numpy as np
```

```python
PATH_VAL_CLIM = 'Valores_Climatologicos_1970_2024.csv'
```

```python
dtypes_dict = {
    #'fecha': str,
    'indicativo': str,
    'nombre': str,
    'provincia': str,
    'altitud': float,
    'tmed': float,
    'prec': float,
    'tmin': float,
    'tmax': float,
    'velmedia': float,
    'sol': float,
    'presMax': float,
    #'horaPresMax': str,
    'presMin': float,
    # 'horaPresMin': str,
    'hrMedia': float,
    # 'horatmin': str,
    # 'horatmax': str,
    'dir': float,
    'racha': float,
    # 'horaracha': str,
    'hrMax': float,
    # 'horaHrMax': str,
    'hrMin': float,
    # 'horaHrMin': str
}
```

```python
COLS_NOMBRES = [
    'indicativo',
    'nombre',
    'provincia',
]

COLS_NUMS = [
    'altitud',
    'tmed',
    'prec',
    'tmin',
    'tmax',
    'velmedia',
    'sol',
    'presMax',
    'presMin',
    'hrMedia',
    'dir',
    'racha',
    'hrMax',
    'hrMin',
]

COL_FECHA = 'fecha'
COLS_HORAS = [
    'horaPresMax',
    'horaPresMin',
    'horatmin',
    'horatmax',
    'horaracha',
    'horaHrMax',
    'horaHrMin'
]
```

```python
# dtypes_dict = {nombre: str for nombre in COLS_NOMBRES} | {num: float for num in COLS_NUMS}
```

```python
df = pd.read_csv(PATH_VAL_CLIM,
                 sep=';', decimal = ',',
                 dtypes = {nombre: str for nombre in COLS_NOMBRES},
                 parse_dates = ['fecha'],
                 low_memory = False)
```

```python
display(df)
```

```python
df.dtypes
```

```python
cols_nums_mal = [col for col in COLS_NUMS if df.dtypes[col] == object]
print(cols_nums_mal)
```

```python
# Enumeración de las columnas, sus tipos y sus posibles valores.
print('\nColumnas:')
for column in cols_nums_mal:
    values = df[column].unique()
    dtype = df.dtypes[column]
    print(column + ':', dtype, values)

# Valores nulos en cada columna.
# print('\nValores nulos:')
# print(df.isnull().sum())
```

```python
COL_PREC = 'prec'
cols_nums_mal.remove(COL_PREC)
```

```python
df_prep = df.copy(deep = True)
```

```python
df_prep[cols_nums_mal] = df_prep[cols_nums_mal].apply(pd.to_numeric, errors = 'raise')
```

```python
df_prep.dtypes
```

```python
df_prep['prec'] = pd.to_numeric(df_prep['prec'], errors='raise')
```

```python
df_prep[COL_PREC] = pd.to_numeric(df_prep[COL_PREC].str.replace(',', '.'), errors='raise')
```
