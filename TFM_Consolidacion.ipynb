```python
import pandas as pd
import os
```

```python
PATH_DATOS = 'Datos'
FORMATO = '.csv'
```

```python
# Listamos todos los CSVs
archivos = [archivo for archivo in os.listdir(PATH_DATOS) if archivo.endswith(FORMATO)]
```

```python
df_con = pd.DataFrame()

for archivo in archivos:
        
        ruta_archivo = os.path.join(PATH_DATOS, archivo)
        print(f"Procesando {archivo}")
        df = pd.read_csv(ruta_archivo, sep=';')
        df_con = pd.concat([df_con, df], axis=0, ignore_index=True)
        print(f"EXITO: {archivo} procesado correctamente.")
```

```python
display(df_con)
```

```python
df_con.drop(columns=df_con.columns[0], axis=1, inplace=True)
```

```python
display(df_con)
```

```python
df_con.to_csv('Valores_Climatologicos_1970_2024.csv', sep = ';', index = False)
```
